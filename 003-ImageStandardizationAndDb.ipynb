```python
dir_dic = [
'images/OnlineDatabases/DatasetA/MSKCC',
'images/OnlineDatabases/DatasetA/BBBC008_v1_images',
'images/OnlineDatabases/DatasetA/BBBC018_v1_images' ,
'images/OnlineDatabases/DatasetA/nucleussegdata/nucleussegdata/TrainingSet',
'images/OnlineDatabases/DatasetA/nucleussegdata/nucleussegdata/HepG2TestSet',
'images/OnlineDatabases/DatasetA/nucleussegdata/nucleussegdata/Huh7TestSet',
'images/OnlineDatabases/DatasetA/cell_image_library']

cols = ['id', 'size_Y', 'size_X', 'mean_intensity', 'std_intensity', ]
stats = pd.DataFrame(columns = cols)
```

```python
for dir_temp in dir_dic:
    for image in os.listdir(dir_temp):
        print(image)
        ext = str(image[-4:-1]+image[-1])
        if ext == ".jpg" or ext == ".DIB" or ext == ".tif":
            
            images_loc = str("./"+dir_temp+"/"+image)
            im_temp = Image.open(images_loc).convert('L')
            
            stats_image = {"id": [image], "size_Y":[float(im_temp.size[1])],"size_X":float(im_temp.size[0]) , "mean_intensity": [float(np.mean(im_temp))], "std_intensity": [float(np.std(im_temp))]}
            
            list_stats = [[image, float(im_temp.size[1]), float(im_temp.size[0]), float(np.mean(im_temp)), float(np.std(im_temp))]]
            
            #print(stats_image)
            #list_temp = [image, float(im_temp.size[1]), float(np.mean(im_temp)), float(np.std(im_temp))]
            
            temp_db = pd.DataFrame(list_stats, columns = cols)
            stats = stats.append(temp_db, ignore_index=True)
            im_temp = np.array(im_temp)
            
            im_temp = Image.fromarray((im_temp - np.mean(im_temp) / np.std(im_temp)).astype(np.uint8), 'L')
            im_temp.save(str("StandardizedImages/WholeTissue/"+image+"standard.png"), "PNG")
```
