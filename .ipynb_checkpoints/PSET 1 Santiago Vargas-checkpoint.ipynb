In order to properly span the space of possible functions, I used scipy.curve_fit. This function allows for a user to define the class of functions they which to optimize over. This will allow us to look beyond a polynomial function. 

```python
def func(x, a1, a2, a3, a4, a5, a6, a7, a8, a9, a10, a11, a12, a13, a14, a15, a16 ):
    tau = a1
    return a2 * np.sin(1 * np.pi / tau * x) + \
           a3 * np.sin(2 * np.pi / tau * x) + \
           a4 * np.sin(3 * np.pi / tau * x) + \
           a5 * np.sin(4 * np.pi / tau * x) + \
           a6 * np.sin(5 * np.pi / tau * x) + \
           a7 * np.sin(6 * np.pi / tau * x) + \
           a8 * np.sin(7 * np.pi / tau * x) + \
           a9 * np.sin(8 * np.pi / tau * x) + \
           a10 * np.sin(9 * np.pi / tau * x) + \
           a11 * np.sin(10 * np.pi / tau * x) + \
           a12 * np.sin(11 * np.pi / tau * x) + \
           a13 * np.sin(12 * np.pi / tau * x) + \
           a14 * np.sin(13 * np.pi / tau * x) + \
           a15 * np.sin(14 * np.pi / tau * x) + \
           a16 * np.sin(15 * np.pi / tau * x) 
```

```python
def fourier(x, *a):
    tau = a[0]
    ret = a[1] * np.sin(np.pi / tau * x)
    for deg in range(2, len(a)):
        ret += a[deg] * np.sin((deg+1) * np.pi / tau * x)
    return ret
```

```python
x3_slice = data3[:,0]
y3_slice = data3[:,1]

from scipy.optimize import curve_fit
#data is antisymmeric so lets only use odd fourier coeffecients

#1: test with polynomials. Goal: beat polynomial with fourier.
z = np.polyfit(x3_slice, y3_slice,15)
poly = np.poly1d(z)
y_custom = poly(x3_slice)

#2: test with more custom function
opt2, cov = curve_fit(func, x3_slice, y3_slice)
y_fourier = func(x3_slice, *opt2)


#3: test with more generalized fourier function

four_step = 35
a = [1.5]
for i in range(four_step):
    a = a + [np.power(-1,i+1)]

opt3, cov = curve_fit(fourier, x3_slice, y3_slice, a)
y_custom_fourier = fourier(x3_slice, *opt3)

```

```python
#plot

fig , axs = plt.subplots(3,1)
fig.set_figheight(30)
fig.set_figwidth(10)

temp = "optimization"
axs[0].set_title("15-term polynomial fit Optimization")
axs[0].plot(x3_slice, y_custom, 'r-', label = temp)
axs[0].scatter(x3_slice,y3_slice, s = 1, label= "data")
axs[0].legend()

axs[1].set_title("15-term Fourier Function Optimization")
axs[1].plot(x3_slice, y_fourier, 'r-', label = temp)
axs[1].scatter(x3_slice,y3_slice, s = 1, label= "data")
axs[1].legend()

temp ="fit"
axs[2].set_title("Custom Fourier Function Optimization")
axs[2].plot(x3_slice, y_custom_fourier, 'r-', label = temp)
axs[2].scatter(x3_slice,y3_slice, s = 1, label= "data")
axs[2].legend()

```

There are two solid guesses for what the function might be, a polynomial and fourier-based guess. 


The Fourier based function has the form

$$ \Sigma ^{30}_{n = 1} a_i sin( \frac{\pi}{\tau} x ) $$

from the optimization above $$\tau = 1.61 $$ for all of the transforms


The polynomial takes the form 

$$ \Sigma ^{15}_{n = 1} a_i x^i $$

Problem 3 A

The accuracy of the tensorflow code denotes the degree to what proportion of the images in a given set are correctly captured by the model. For example the accuracy on the training data indicates what proportion of the images are correctl encompassed in the model. 

This differs from the accuracy of each individual data point in the set. For this data, the accuracy is the confidence that the image is a certain item. 

```python
import tensorflow as tf 
from tensorflow import keras
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
print(np.size(train_images))
print(np.size(train_labels))
print(train_labels[1:5])
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

The first layer in this network, tf.keras.layers.Flatten, transforms the format of the images from a 2d-array (of 28 by 28 pixels), to a 1d-array of 28 * 28 = 784 pixels.

After the pixels are flattened, the network consists of a sequence of two tf.keras.layers.Dense layers. These are densely-connected, or fully-connected, neural layers

Softmax layer of the size of the output array gives probability of each data point being the given data

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)

print('Test accuracy:', test_acc)
```

```python
predictions = model.predict(test_images)
```

```python
from sklearn.decomposition import PCA

i = 8
#train_images, train_labels), (test_images, test_labels) 
pca = PCA(n_components = i, whiten = True)

num = train_images.shape
print(num)
resize = np.resize(train_images,(60000,784))
resize_test = np.resize(test_images,(10000,784))
print(resize.shape)
X_train = pca.fit_transform(resize)
X_test = pca.transform(resize_test)
```

```python
print(pca.explained_variance_ratio_)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(solver="lbfgs")
classifier.fit(X_train,train_labels)
pred = classifier.predict(X_test)
```
